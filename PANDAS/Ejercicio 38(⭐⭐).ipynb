#Primero realizamos un filtrado para arreglar la informacion de las categorias

```python
filtrado_inicial = infoplay[['App','Category']]
filtrado_inicial['Category'] = infoplay['Category'].str.replace('_',' ')
filtrado_inicial["Category"] = filtrado_inicial["Category"].explode("Category")
```

```python
filtrado_inicial['Category'].unique()
```

```python
filtrado_inicial = filtrado_inicial.dropna(subset='Category')
filtrado_inicial = filtrado_inicial[filtrado_inicial['Category'] != '1.9']
```

```python
filtrado_inicial['Category'].unique()
```

```python
def filtrar_caracteres(texto):
    return ''.join(filter(str.isalpha, texto))
```

```python
filtrado_inicial = filtrado_inicial.dropna(subset='App')
```

Nos quedamos solo con las letras de la A a la Z (mayúsculas y minúsculas)

```python
filtrado_inicial['App'] = filtrado_inicial['App'].apply(filtrar_caracteres)
filtrado_inicial['App'] = filtrado_inicial['App'].map(lambda x: x.strip())
```

Vamos a separar a todas las app por categorias

```python
reorganizacion = filtrado_inicial.pivot(columns='Category', values='App')
reorganizacion = reorganizacion.fillna(False)
```

```python
reorganizacion
```

#Aplicamos una funcion que lleve un contador de la letra predominante en cada nombre de app para una categoria

```python
def encontrar_caracter_predominante(texto):
    if texto is not False:
        contador_char = Counter(texto)
        if contador_char:
            char_mas_frecuente = contador_char.most_common(1)[0][0]
            return char_mas_frecuente
    return False
```

Obtengo una serie que contiene el caracter mas frecuente de cada elemento de la app ART AND DESIGN

```python
lista_mejor_caracter = reorganizacion['ART AND DESIGN'].map(encontrar_caracter_predominante)
```

```python
serie_filtrada = lista_mejor_caracter.loc[lista_mejor_caracter != False]
```

```python
serie_filtrada
```

```python
frecuencia_caracteres = serie_filtrada.value_counts()
caracter_mas_comun = frecuencia_caracteres.idxmax()
```

```python
caracter_mas_comun
```

#Ahora tengo que hacer esto para cada categoria

```python
def encontrar_caracter_predominante(texto):
    if texto is not False:
        contador_char = Counter(texto)
        if contador_char:
            char_mas_frecuente = contador_char.most_common(1)[0][0]
            return char_mas_frecuente
    return None
```

```python
resultados = pd.DataFrame()
for columna in reorganizacion.columns:
    if columna:
        resultados[columna] = reorganizacion[columna].apply(encontrar_caracter_predominante)
```

```python
caracteres_predominantes = resultados.mode().iloc[0]
caracteres_predominantes
```
