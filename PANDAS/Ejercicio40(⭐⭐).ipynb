#Creo una funcion que con el uso de una expresion regular, detecta el numero dentro del string y luego lo multiplica por el valor necesario en caso de encontrar una K o una M

(Interprete a los K como 1.000 de bytes y a los M como 1.000.000 de bytes. Los resultados se expresan en Bytes)

```python
def procesar_valor(texto):
    parte_numerica = re.search(r'\d+(\.\d+)?', texto)
    if parte_numerica:
        valor_numerico = float(parte_numerica.group())
        if 'M' in texto:
            valor_numerico *= 1000000
        elif 'k' in texto:
            valor_numerico *= 1000
        return valor_numerico
    else:
        return None
```

```python
filtrado_inicial['Size'] = filtrado_inicial['Size'].map(lambda x: procesar_valor(x))
```

```python
filtrado_inicial['Size'] = pd.to_numeric(filtrado_inicial["Size"], errors='coerce')
filtrado_inicial = filtrado_inicial.dropna(subset=['Size'])
```

```python
resolucion = filtrado_inicial.groupby('Genres')['Size'].sum().reset_index()
```

Hice las operaciones necesarias en Bytes pero para que sea mas clara la visualizacion de la respuesta paso todo a GB

```python
resolucion['Size'] = resolucion['Size'].map(lambda x: x/1000000000)
```
