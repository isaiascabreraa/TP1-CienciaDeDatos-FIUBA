```python
aplicaciones_educativas = infoplay[['App','Genres','Size']]
```

```python
aplicaciones_educativas = aplicaciones_educativas.dropna(subset='Genres')
aplicaciones_educativas["Genres"] = aplicaciones_educativas["Genres"].map(lambda x: x.split(';')).explode('Genres')
aplicaciones_educativas = aplicaciones_educativas[aplicaciones_educativas['Genres'] == 'Education']
```

```python
aplicaciones_educativas = aplicaciones_educativas[aplicaciones_educativas['Size'] != 'Varies with device']
```
