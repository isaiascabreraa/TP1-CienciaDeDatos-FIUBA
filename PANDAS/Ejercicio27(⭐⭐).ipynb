```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
```

```python
infoplay = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TP1-VISU/googleplaystore.csv")
inforeviews = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/TP1-VISU/googleplaystore_user_reviews.csv")
```

```python
merged_info = pd.merge(infoplay, inforeviews, on='App', how='inner')
merged_info
```

```python
consulta_NPL = merged_info[['App','Rating','Translated_Review']]
```

```python
consulta_NPL = consulta_NPL.dropna()
```

```python
consulta_NPL
```

Uso TfidfVectorizer, que es una clase que se utiliza para convertir texto en un formato numérico que representa la importancia relativa de las palabras en un conjunto de documentos. Este objeto me permite configurarlo de cierta forma como lowercase=True, que convierte todas las palabras a minúsculas durante el procesamiento.

```python
vector_features = TfidfVectorizer(lowercase=True, stop_words='english', max_features=10000)
vector_features
```

Tomo un objeto TfidfVectorizer previamente configurado, que es vector_features en este caso, y lo ajusta (fit) a un conjunto de datos específico

```python
matrix_features = vector_features.fit_transform(consulta_NPL['Translated_Review'])
matrix_features
```

```python
nbrs = NearestNeighbors(metric='cosine', algorithm='brute')
nbrs.fit(matrix_features)
```

```python
busqueda = vector_features.transform(["love game"])
```

```python
distances, indexes = nbrs.kneighbors(busqueda, n_neighbors=1)
```
