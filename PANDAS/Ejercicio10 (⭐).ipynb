```python
promedio_por_opinion = inforeviews[['Translated_Review','Sentiment']]
promedio_por_opinion = promedio_por_opinion.dropna(subset='Sentiment').dropna(subset='Translated_Review')
```

Nos quedamos solo con las letras de la A a la Z (mayúsculas y minúsculas)

```python

def filtrar_caracteres(texto):
    return ''.join(filter(str.isalpha, texto))
```

```python
promedio_por_opinion['Translated_Review'] = promedio_por_opinion['Translated_Review'].apply(filtrar_caracteres)
promedio_por_opinion['Translated_Review'].unique()
```

Calcula el largo promedio de las reseñas por tipo de opinión y las redondea a un entero

```python
promedio_por_opinion['Largo_Review'] = promedio_por_opinion['Translated_Review'].apply(len)
promedio_por_opinion = promedio_por_opinion.groupby('Sentiment')['Largo_Review'].mean().round().astype(int)

print(promedio_por_opinion)
```
