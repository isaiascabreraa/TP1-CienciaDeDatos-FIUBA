```python
sqlContext = SQLContext(sc)
infoplay = sqlContext.read.csv("/content/drive/MyDrive/Colab Notebooks/TP1-VISU/googleplaystore.csv", header=True)
```

```python
appinfoplayRDD = infoplay.rdd
```

```python
appinfoplayRDD.take(10)
```

Muestro las aplicaciones con su tamaño y genero correspondiente. Solo tomo 20 delementos porque de hacer un collect se dificultaria la visualizacion por la cantidad que son.

```python
appinfoplayRDD.filter(lambda x: x.Genres == 'Education' and all(elemento is not None for elemento in x)).map(lambda x: (x.App, (x.Size, x.Genres))).distinct().take(20)
```
