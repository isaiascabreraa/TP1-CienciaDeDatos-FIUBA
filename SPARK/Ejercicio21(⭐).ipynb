```python
# Crea la sesion de SPARK
spark = SparkSession.builder.getOrCreate()

# Crea el SPARK content
sc = spark.sparkContext
```

```python
sqlContext = SQLContext(sc)
infoplay = sqlContext.read.csv("/content/drive/MyDrive/Colab Notebooks/TP1-VISU/googleplaystore.csv", header=True)
inforeviews = sqlContext.read.csv("/content/drive/MyDrive/Colab Notebooks/TP1-VISU/googleplaystore_user_reviews.csv", header=True)
```

```python
appinfoplayRDD = infoplay.rdd
appinforeviewsRDD = inforeviews.rdd
```

Reemplazo los espacios en blanco entre los nombres de las app

```python
def eliminarEspacios(cadena):
  return cadena.replace(' ', '')
```

Tranformo mi RDD en pares clave-valor donde se indica el nombre de la App y la cantidad de letras que esta ocupa. Luego uno a ambos RDD y me quedo el que tiene mas letras.

```python
appinforeviewsRDD.map(lambda x: (eliminarEspacios(x.App), len(eliminarEspacios(x.App)) )).union(
    appinfoplayRDD.map(lambda x:(eliminarEspacios(x.App), len(eliminarEspacios(x.App)) ))).filter(lambda x: all(elemento is not None for elemento in x)).distinct().reduce(lambda x,y: x if x[1] > y[1] else y)
```
