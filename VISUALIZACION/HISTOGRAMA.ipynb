```python
#Chequeamos que la escala sea entre -1 y 1 (y que sean numeros)
#Se dropean las filas que tienen nan porque resultan inconsistentes para nuestro analisis ya que no sabemos que valor deben contener. Completarlas aleatoriamente unicamente quitaria veracidad a nuestro analisis
inforeviews = inforeviews.dropna(subset=['Sentiment_Polarity'])
inforeviews['Sentiment_Polarity'].describe()
```

```python
inforeviews['Sentiment'].unique()
```

```python
inforeviews = inforeviews.dropna(subset=['Sentiment'])
```

```python
sentpolarity = inforeviews[['Sentiment','Sentiment_Polarity']]
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 5))
sns.histplot(data=sentpolarity, x="Sentiment_Polarity", bins=40, color='#0055A4')

plt.title('Distribución de opiniones en reseñas de Usuarios', pad=10)
plt.xlabel('Polaridad de opiniones', labelpad=10)
plt.ylabel('Frecuencia de aparición', labelpad=10)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
