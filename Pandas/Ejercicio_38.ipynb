```python
# 38) Para cada categoría, calcular cuál es el caracter predominante en la misma. El caracter predominante en una categoria esta dado por el caracter que es
# dominante en mayor número de apps, por ej para aaaaaaax, xxb, xxs el caracter dominante de ap1 es a, pero para las otras 2 x, por lo que x es el caracter
# dominante en la categoría. Si no hay caracter dominante, es el primer caracter. (⭐⭐)
```

```python
#Primero arreglamos la informacion de las categorias
ejercicio40 = infoplay[['App','Category']]
ejercicio40['Category'] = infoplay['Category'].str.replace('_',' ')
ejercicio40["Category"] = ejercicio40["Category"].explode("Category")
```

```python
ejercicio40['Category'].unique()
```

```python
ejercicio40 = ejercicio40.dropna(subset='Category')
ejercicio40 = ejercicio40[ejercicio40['Category'] != '1.9']
```

```python
ejercicio40['Category'].unique()
```

```python
#Nos quedamos solo con las letras de la A a la Z (mayúsculas y minúsculas)
def filtrar_caracteres(texto):
    return ''.join(filter(str.isalpha, texto))
```

```python
ejercicio40 = ejercicio40.dropna(subset='App')
```

```python
#Aplicamos un filtro para las letras y los espacios en blanco
ejercicio40['App'] = ejercicio40['App'].apply(filtrar_caracteres)
ejercicio40['App'] = ejercicio40['App'].map(lambda x: x.strip())
```

```python
#Vamos a separar a todas las app por categorias
reorganizacion = ejercicio40.pivot(columns='Category', values='App')
reorganizacion = reorganizacion.fillna(False)
```

```python
reorganizacion
```

```python
#Aplicamos una funcion que lleve un contador de la letra predominante en cada nombre de app para una categoria
def encontrar_caracter_predominante(texto):
  if texto is not False:
    contador_char = Counter(texto)
    contador_char = contador_char.most_common()
    char_mas_frecuente = contador_char[0][1]
    char_predominante = [char for char, count in contador_char if count == char_mas_frecuente]
    return char_predominante[0]  # Devuelve el primer carácter predominante
  else:
    return False
```

```python
#Obtengo una serie que contiene el caracter mas frecuente de cada elemento de la app ART AND DESIGN
lista_mejor_caracter = reorganizacion['ART AND DESIGN'].map(encontrar_caracter_predominante)
```

```python
serie_filtrada = lista_mejor_caracter.loc[lista_mejor_caracter != False]
```

```python
serie_filtrada
```

```python
frecuencia_caracteres = serie_filtrada.value_counts()
caracter_mas_comun = frecuencia_caracteres.idxmax()
```

```python
frecuencia_caracteres
```

```python
caracter_mas_comun
```

```python
#Ahora tengo que hacer esto para cada categoria
```

```python
resultados = pd.DataFrame()

for columna in reorganizacion.columns:
    if columna:
      resultados[columna] = reorganizacion[columna].map(encontrar_caracter_predominante)

      serie_filtrada = resultados[columna].loc[resultados[columna] != False]
      frecuencia_caracteres = serie_filtrada.value_counts()
      caracter_mas_comun = frecuencia_caracteres.idxmax()

      resultados[columna] = caracter_mas_comun
```

```python
resultados = resultados.T
```

```python
resultados
```

```python
resultados = resultados.rename(columns={0: 'Caracter mas frecuente'})
resolucion_ejercicio40 = resultados['Caracter mas frecuente']
```
