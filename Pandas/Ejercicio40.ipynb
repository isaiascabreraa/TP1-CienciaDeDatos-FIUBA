```python
dfinicial = infoplay[['App','Genres','Size']]
```

```python
dfinicial['Genres'] = dfinicial['Genres'].map(lambda x: x.split(';'))
```

```python
dfinicial = dfinicial.explode('Genres')
```

```python
dfinicial
```

```python
dfinicial['Size'].unique()
```

```python
#Interprete a los K como 1.000 y a los M como 1.000.000. Los resultados se expresan en Bytes
dfinicial['Size'] = dfinicial['Size'].str.replace('k','000')
dfinicial['Size'] = dfinicial['Size'].str.replace('M','000000')
```

```python
dfinicial['Size'] = pd.to_numeric(dfinicial["Size"], errors='coerce')
dfinicial = dfinicial.dropna(subset=['Size'])
```

```python
dffinal = dfinicial.groupby('Genres')['Size'].sum().reset_index()
```
